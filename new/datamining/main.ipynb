```python
data = pd.merge(data0, data2)
data
```

```python
data = data[data.day > data.birthday]
data
```

```python
data['age'] = round(( data['day'] - data['birthday'] ) / pd.Timedelta(days = 365) , 2)
data = data[data.age < 5]
data.age.describe()
```

```python
columns_to_drop = ['user_id', 'auction_id', 'day', 'birthday', 'month', 'birthmonth']
data = data.drop(columns=columns_to_drop)
data
print(data.columns)
```

```python
from sklearn.cluster import KMeans
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split


features = ['category_2', 'category_1', 'buy_mount', 'year', 'quarter', 'gender', 'birthyear', 'birthquarter', 'age']
X = data[features]

# 创建KMeans模型并训练
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)

# 获取聚类结果标签
labels = kmeans.labels_

# 将聚类结果添加到原始DataFrame
data['cluster_label'] = labels

data
```

```python
from sklearn.metrics import accuracy_score


# 划分训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(X, data['cluster_label'], test_size=0.2, random_state=42)

# 创建分类器模型并训练
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)

# 对所有样本进行预测
data['predicted_target'] = clf.predict(X)
y_pred = clf.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("分类准确率：", accuracy)
```
